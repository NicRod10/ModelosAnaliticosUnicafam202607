```python
# Función para determinar si una persona es mayor o menor de edad

def verificar_edad(edad):
    if edad >= 18:
        return "La persona es mayor de edad."
    else:
        return "La persona es menor de edad."

# Solicitar la edad al usuario
edad = int(input("Ingrese la edad de la persona:"))

# Mostrar el resultado
print(verificar_edad(edad))
```

```python
# Función para clasificar una nota

def clasificar_nota(nota):
    if nota >= 90 and nota <= 100:
        return "Excelente"
    elif nota >= 70:
        return "Bueno"
    elif nota >= 50:
        return "Suficiente"
    elif nota >= 0:
        return "Insuficiente"
    else:
        return "La nota ingresada no es válida."

# Solicitar la nota al usuario
nota = int(input("Ingrese la nota (0 a 100): "))

# Mostrar el resultado
print("Clasificación:", clasificar_nota(nota))
```

```python
# Función para verificar la contraseña

def verificar_contraseña(contraseña):
    contraseña_correcta = "secreta123"

    if contraseña == contraseña_correcta:
        return "La contraseña es correcta."
    else:
        return "La contraseña es incorrecta."

# Solicitar la contraseña al usuario
contraseña = input("Ingrese la contraseña: ")

# Mostrar el resultado
print(verificar_contraseña(contraseña))
```

```python
# Función para calcular el IMC y retornar la categoría

def imc(peso, altura):
    valor_imc = peso / (altura ** 2)

    if valor_imc < 18.5:
        categoria = 'Bajo peso'
    elif valor_imc < 25:
        categoria = 'Normal'
    elif valor_imc < 30:
        categoria = 'Sobrepeso'
    else:
        categoria = 'Obesidad'

    return round(valor_imc, 2), categoria

# Solicitar datos al usuario
peso = float(input('Ingrese su peso en kg: '))
altura = float(input('Ingrese su altura en metros: '))

# Obtener resultados
resultado_imc, categoria = imc(peso, altura)

# Mostrar resultados
print('IMC:', resultado_imc)
print('Categoría:', categoria)
```
